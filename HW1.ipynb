## Part 4

We are going to use `identity_matrix` to scramble the rows and columns of `img_average_channel`.

## Part 5

1. Use your array of random row indices to create a new matrix `T` from `identity_matrix`. Overall, `T` has the same elements as `identity_matrix`, but the order of the rows of `T` should be according to the variable `rows`
2. Take a moment to think about what we are doing here. Describe in your own words what matrix `T` is.
3. Sanity check, print the shape of `T` and make sure it is the same size as `identity_matrix` and `img_average_channel`. If you need to,  reshape it to have the same size.


```python
T = identity_matrix[rows]
# What we're doing here with matrix T is essentially permuting the rows of the identity matrix based on the random order specified by the rows variable. In other words, T is a rearranged identity matrix where the rows are shuffled, and each row retains its original identity (each row is still an identity row), but they are ordered differently. It's like reordering a set of numbered cards.

# Print the shape of T
T.shape
```

## Part 6

3. Think about what each dot product in the matrix multiplication is doing. Describe in your own words what this transformation is doing to the original image. 

```python
# Matrix multiply T and img_average_channel
transformed_image = T @ img_average_channel

# Print shape of transformed image
transformed_image.shape
```

```python
# Plot transformed_image
plt.axis('off')
plt.imshow(transformed_image)

# The transformation performed by matrix multiplication with T reorders the rows of the original image,
# creating a new arrangement of the image data while retaining the same values.
```

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python

# Calculate the determinant of T and check if it is nonzero.
determinant = np.linalg.det(T)
is_invertible = determinant != 0

if is_invertible:
    print("Matrix T is invertible.")
    Tinv = np.linalg.inv(T)

    # Apply the inverse transformation to the transformed image
    retrieved_image = Tinv @ transformed_image

    # Reshape the retrieved_image to match the original img_average_channel shape
    retrieved_image = retrieved_image.reshape(img_average_channel.shape)

    # Plot the retrieved image to show that you retrieved the original image.
    fig = plt.figure(figsize=(15, 10))

    # Plot the original image 
    plt.subplot(1, 2, 1)
    plt.axis('off')
    plt.imshow(retrieved_image)
    plt.title("Retrieved Image")

    plt.subplot(1, 2, 2)
    plt.axis('off')
    plt.imshow(img_average_channel)
    plt.title("Original Image")

else:
    print("Matrix T is not invertible.")
```

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image.

```python
# Create an aray using the image
image = np.array(Image.open('lemon2.jpeg'))

# Get height, width and channels of the image
height, width, channels = image.shape

# Create and identity matrice for both rows and columns
identity_rows = np.identity(height)
identity_cols = np.identity(width)

# Shuffle both row and column indices
row_indices = np.arange(height)
col_indices = np.arange(width)

np.random.shuffle(row_indices)
np.random.shuffle(col_indices)


# Create a variable with same dimensions as image
scrambled_image = np.zeros_like(image)

# Create matrices T_rows and T_cols for row and column shuffling
T_rows = identity_rows[row_indices, :]
T_cols = identity_cols[:, col_indices]

# Scramble the channels 
for channel in range(channels):
    scrambled_image[:, :, channel] = T_rows @ image[:, :, channel] @ T_cols

# Plotting scrambled image
plt.imshow(scrambled_image)
plt.axis('off')
plt.title("Scrambled Image")
plt.show()

T_cols_inv = np.linalg.inv(T_cols)

T_rows_inv = np.linalg.inv(T_rows)

# Create a variable with same dimensions as image
recovered_image = np.zeros_like(image)

# Retrieve  original image 
for channel in range(channels):
    recovered_image[:, :, channel] = T_rows_inv @ scrambled_image[:, :, channel] @ T_cols_inv

# Plot the retrieved original image
plt.imshow(recovered_image)
plt.title("Retrieved Original Image")
plt.axis('off')
plt.show()
```
